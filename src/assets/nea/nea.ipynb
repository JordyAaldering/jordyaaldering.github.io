```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid', rc={'ytick.left': True, 'ytick.color': 'silver'})
plt.rcParams['text.usetex'] = False
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['axes.autolimit_mode'] = 'round_numbers'
plt.rcParams['ytick.labelcolor'] = 'black'
sns.set_palette('colorblind')
```

```python
df2 = pd.read_csv('nea.csv')
df1 = df2[df2['Language'] != 'Rust Tokio (alloc)']
```

```python
def plot(df, y):
    sns.barplot(df, x='Benchmark', y=y, hue='Language', errorbar='ci')
    plt.xlabel(None)
    plt.tight_layout()
```

```python
plot(df1, y='Runtime (s)')
plt.ylim(0, 2.5)
plt.legend(loc='lower left')
plt.savefig('runtime.svg', bbox_inches='tight', transparent=True)
```

```python
plot(df1, y='INA system energy (J)')
plt.legend().set_visible(False)
plt.savefig('energy.svg', bbox_inches='tight', transparent=True)
```

```python
plot(df2, y='Runtime (s)')
plt.ylim(0, 2.5)
plt.legend(loc='lower left')
plt.savefig('alloc_runtime.svg', bbox_inches='tight', transparent=True)
```

```python
plot(df2, y='INA system energy (J)')
plt.legend().set_visible(False)
plt.savefig('alloc_energy.svg', bbox_inches='tight', transparent=True)
```
